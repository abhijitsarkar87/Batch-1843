# HISTOGRAM

Histogram takes in a series of data and divides the data into a number of bins. It then plots the frequency data points in each bins (as per interval).

Work?

>. Helps in counting or understanding the count of data ranges.

When to use?

>. We should use histogram when we need the count of the variable in a plot.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a= np.array([22,45,54,37,86,67,34,12,29,47,78])

plt.hist(a,bins=[0,20,40,60,80,100])
plt.title('Histogram')
plt.show()
```

```python
# Observation:

# here 0 to 20 is 1, 
# 20 to 40 is 4,
# 40 to 60 is 3.... respectively
```

```python
import pandas as pd
df=pd.read_csv("student_marks.csv",
               index_col='Name',
               parse_dates=['DOB'])
```

```python
plt.hist(df['Maths'],bins=10)
plt.title("Maths Marks Histogram")
plt.show()
```

```python
plt.hist(df['English'],bins=10)
plt.title('English Marks Histogram')
plt.show()
```

```python
plt.hist(df['Biology'],bins=10)
plt.title("Biology Marks Histogram")
plt.show()
```

```python
plt.hist(df['Chemistry'],bins=10, align='mid')
plt.title("Chemistry Marks Historam")
plt.show()
```
