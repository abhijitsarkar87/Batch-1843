## IRIS FLOWER DATASET:

```python
ds=pd.read_csv("iris.csv")
```

```python
ds
```

```python
# one numerical and one class(By category)

sns.set(style="whitegrid") #background style
ax=sns.stripplot(x='class',y='sepal length',data=ds)
plt.title("Iris category Plot")
plt.show()
```

```python
sns.set(style="whitegrid") #background
# x is class and y is sepel width.
ax=sns.stripplot(x='class',y='sepal width',data=ds)
plt.title("Iris category Plot by width")
plt.show()
```

```python
sns.swarmplot(x='class',y='sepal length',data=ds)
plt.title("Category Plot")
plt.show()
```

```python
sns.swarmplot(x='petal length',y='class',data=ds)
plt.title("Category PLot ")
plt.show()
```

```python
#counting by class:
ds['class'].value_counts()
```

# Distribution Plot:

>displot: Figure-level interface to distribution plot functions.

>histplot   : Plot a histogram of binned counts with optional normalization or smoothing.

>jointplot  : Draw a bivariate plot with univariate marginal distributions.

>violinplot : Draw an enhanced boxplot using kernal density estimation.

>rugplot: Plot marginal distributions by drawing ticks along the x and y axes.  

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.violinplot(x='total_bill',data=tips)
plt.show
```

```python
#sns.histplot?
```

```python
sns.histplot(data=tips,x='total_bill',bins=20)
```

```python
sns.distplot(tips['tip'],bins=10)
```

```python
sns.distplot(tips["total_bill"])
```

```python
#sns.histplot?
```

```python
sns.histplot(data=tips,x="total_bill",bins=10)
```

```python
sns.histplot(data=tips,x="total_bill",bins=100)
```

```python
# for distplot. KDE=kernal density estimation.

sns.displot(tips['total_bill'],bins=50,kde=True)
```

```python
sns.displot(tips['total_bill'],bins=50,kde=False) #no KDE
```

```python
#sns.kdeplot?
```

```python
sns.kdeplot(tips["total_bill"])
```

#### COMPOSITE VIEWS ONTO MULTIVARIATE DATASETS.

```python
sns.jointplot(data=tips,x='total_bill',y='tip') 

# bill increases then tips also gets increased, 02-04 tips are max 
```

```python
sns.rugplot(data=tips,x='total_bill',y='tip') 
# rugplot shows max amount tips paid is between 2 to 4 
```

### Categorical Distribution Plot

The first is the familier boxplot().This kind of plot shows the three quartile values of the distribution along with extreme values. The "whiskers" extend to points that lie within 1.5 IQrs of the lower and upper quartil, and the observations that fall outside the range are displayed independently.This means that each value in the boxplot corresponds to an actual observation in the data.

```python
sns.boxplot?
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips, palette="coolwarm")
# kind=? parameter is not given in this parameter
```

```python
sns.pairplot(tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips) 
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips)
```

```python
import seaborn as sns
flights=sns.load_dataset("flights")

# pivot will take 3 parameter(month,year, passenger boarded)

flights=flights.pivot("month","year","passengers")
flights
```
