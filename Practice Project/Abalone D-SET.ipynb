# Checking Multicolinarity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(x):
    vif=pd.DataFrame()
    vif['Variables']=x.columns
    vif['vif Factor']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return vif

calc_vif(x)
```

```python
# here length and Diameter is showing Multicolinarity
# Whole Weight and shucked weight showing Multicolinarity
```

```python
df.corr().Rings # checking for lowest correlation to drop column
```

VIF value above 10 and If two or more than two columns has vif value above 10 then always remove the column which has highest VIF then again check the vif value.

We will drop Diameter as its showing multicolinarity with  Length, where high contribution of length.

```python
x.drop(['Diameter'],axis=1,inplace=True)
```

```python
calc_vif(x)
```

shucked weight is showing less contribution as compare to Whole weight.. we will delete Shucked weight

```python
x.drop(['Whole weight'],axis=1,inplace=True)
```

```python
calc_vif(x)
```

# Spliting the Data Frame for Train, Test

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
x_train,x_test,y_train,y_test=train_test_split(xtrain,ytrain,test_size=20,random_state=42)
```

## Decision Tree Classifier

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()

Para={"criterion":['gini', 'entropy', 'log_loss'],"splitter":["best", "random"],"max_features" : [ 'auto', 'sqrt', 'log2']}

gsv=GridSearchCV(dtc,Para)
gsv.fit(x_train,y_train)
print(gsv.best_params_)
```

```python
dtc=DecisionTreeClassifier(criterion ='gini',max_features= 'sqrt', splitter= 'random')
dtc.fit(x_train,y_train)
dtc_score=dtc.score(x_train,y_train)

pred_train=dtc.predict(x_train)
pred_test=dtc.predict(x_test)

print("Training Accuracy",accuracy_score(y_train,pred_train)," ","Test_Accuracy",accuracy_score(y_test,pred_test)*100)
print("Confusion Matrix\n",confusion_matrix(y_test,pred_test))
print("Classifiaction Report ",classification_report(y_test,pred_test),'\n')
```

`Model Name= DTC Training Accuracy= 1.0   Test_Accuracy= 60.0

# Cross Validation 

```python
pred_dtc=dtc.predict(x_test)
from sklearn.model_selection import cross_val_score

dtc_s=accuracy_score(y_test,pred_dtc)

for i in range(2,10):
    score = cross_val_score(dtc,x,y,cv=i)
    dtc_mean=score.mean()
    
    print("At Cv : ", i)
    print("cross_val_ MEAN score is ",dtc_mean*100,"accuracy_score is ",dtc_s*100,"\n")    
```
